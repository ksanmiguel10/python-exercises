## Use matplotlib to plot the following equation:
y = x**2 - x + 2

Add an anotation for the point 0, 0, the origin.

```python
import matplotlib.pyplot as plt
```

```python
print(plt)
```

```python
x = list(range(-5,6))
y = [(x ** 2 - x + 2) for x in x]
plt.text(0, 4, '(0,0)', fontsize=10)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = x^2 - x + 2')
plt.scatter(x,y)
plt.plot(x,y, ls=':')
plt.show()
```

## Combine the figures you created in the last step into one large figure with 4 subplots.

```python
plt.figure(figsize=(12,8))

plt.suptitle('Re-learning basic algebra...')

x = list(range(0,8))
y = [math.sqrt(x) for x in x]
plt.subplot(221)
plt.plot(x, y)
plt.title('y = âˆšx')

x = list(range(-7,8))
y = [x ** 3 for x in x]
plt.subplot(222)
plt.plot(x, y)
plt.title('y = x**3')

x = list(range(0,8))
y = [math.tan(x) for x in x]
plt.subplot(223)
plt.plot(x, y)
plt.title('y = tan(x)')

x = list(range(-7,8))
y = [2 ** x for x in x]
plt.subplot(224)
plt.plot(x, y)
plt.title('y = 2^x')

plt.show()
```

## Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend.

```python
x1 = range(0,8)
x2 = range(0,8)
x3 = range(-7,8)
x4 = range(-7,8)

tan_x = [math.tan(n) for n in x1]
sqrt_x = [math.sqrt(n) for n in x2]
cubed_x = [n**3 for n in x3]
dbl_x = [2 ** n for n in x4]

plt.figure(figsize=(12, 8)) 

plt.plot(x1, tan_x, c='blue', label='$\ y = tan(x)$')
plt.plot(x2, sqrt_x, c='green', label='y = sqrt(x)')
plt.plot(x3, cubed_x, c='red', label='y = x^3')
plt.plot(x4, dbl_x, c='yellow', label='y = 2^x')

plt.ylim(-50,50)
plt.xlim(-8,8)

plt.title('Overlayed Figures')
plt.legend(loc='lower right')
plt.xlabel('$x$')

plt.show
```
